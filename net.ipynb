```python
import requests
import os
import csv
```

```python

def get(url):
    response = requests.get(url).json()
    return response
```

```python
foods = []
```

```python

file_paths = os.listdir("data")
for file_path in file_paths:
    print(file_path)
    with open("data/" + file_path, "r") as file:
        reader = csv.reader(file)
        # skip the header
        next(reader)
        for row in reader:
            foods.append(row[0])
            foods.append(row[1])
```

```python
len(foods)
```

```python
foods = list(set(foods))
len(foods)
```

```python
def queryFood(food):
    try:
        obj = get("http://api.conceptnet.io/c/en/" + food)
        return obj
    except Exception as e:
        print(e)
        return 0
```

```python
from time import sleep
```

```python
count = 0

for food in foods:
    print(f'querying {food} (found: {count})', end='\r')
    obj = queryFood(food)
    if obj != 0:
        count += 1
    sleep(2)
```
